```python
filename = "input.txt"
file = open(filename)

line = file.readline().strip()

buffer = "-"*(len(line)+2)
seats= [[ch for ch in buffer]]
while line:
        #make the backing array mutable, and add some padding
    line = "-"+line+"-"
    seats.append([ch for ch in line])
    line = file.readline().strip()

seats.append([ch for ch in buffer] )
```
