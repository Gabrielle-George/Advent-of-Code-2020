```python
import copy
```

```python
filename = "input.txt"
file = open(filename)

line = file.readline().strip()

buffer = "-"*(len(line)+2)
seats= [[ch for ch in buffer]]
while line:
        #make the backing array mutable, and add some padding
    line = "-"+line+"-"
    seats.append([ch for ch in line])
    line = file.readline().strip()

seats.append([ch for ch in buffer] )
```

```python
def check_adjacent_seats(seati, seatj, seats):
    counter = 0
    for i in [-1,0,1]:
        for j in [-1,0,1]:
            if not (i == j and i ==0):
                #if occupied
                if (seats[seati+i][seatj+j] == "#"):
                    counter= counter+1
    return counter
```

```python
seats2 = copy.deepcopy(seats)
counter = 0
for i in range(len(seats2)):
    for j in range(len(seats2[i])):
        if seats2[i][j] == 'L':
            seats2[i][j] = '#'
            counter = counter+1
```

```python
prev = copy.deepcopy(seats)
cur = copy.deepcopy(seats2)
counter = 0
while prev != cur:
    counter = counter +1
    print(counter)
    prev = copy.deepcopy(cur)
    for i in cur:
        print(i[:16])
    for i in range(len(seats)):
        for j in range(len(seats[i])):
            if cur[i][j] == "L":
                adj = check_adjacent_seats(i,j,prev)
                if adj == 0:
                    cur[i][j] = "#"
            if cur[i][j] == "#":
                adj = check_adjacent_seats(i,j,prev)
                if adj >3:
                    cur[i][j] = "L"


        
```
