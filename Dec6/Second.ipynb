```python
filename = "input.txt"
file = open(filename)
groups = file.read().split("\n\n")
```

```python
arr = []
for group in groups:
    arr.append((len(set.intersection(*[set(response) for response in group.split("\n")]))))
print(sum(arr))
```

```python
for group in groups:
    print([set(response) for response in group.split("\n")])
```
