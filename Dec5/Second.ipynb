```python
import itertools
filename = "input.txt"
file = open(filename)
line = file.readline()
tickets = []
while line:
    tickets.append(line.replace("\n",""))
    line = file.readline()
    
```

```python
def convert_rows(row):
    cur_half = 64
    row_num = 0
    for i in row:
        row_num = row_num +cur_half*(i=="B")
        cur_half = cur_half/2
    return row_num

def convert_col(col):
    cur_half = 4
    col_num = 0
    for i in col:
        col_num = col_num + cur_half*(i=="R")
        cur_half = cur_half/2
    return col_num

def convert_to_nums(ticket):
    row = ticket[:7]
    col = ticket[7:]
    return (convert_rows(row), convert_col(col))




```
