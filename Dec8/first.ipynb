```python
filename = "input.txt"
file = open(filename)

ops = []
line = file.readline().replace('\n','')
while line:
    ops.append(line)
    line = file.readline().replace('\n',"")
    
```

```python
def recurse(ops):
    visited = []
    SP = 0
    accum = 0
    while SP not in visited:
        if SP > len(ops):
            print("FOUND")
            print(accum)
            return ops
        op, offset = ops[SP].split()
        visited.append(SP)

        print(ops[SP])

        if op == "acc":
            accum = accum + int(offset)
            SP = SP+1
        elif op == "nop":
            print("nop")
            SP = SP+1
        elif op == "jmp":
            print("jmp")
            SP = SP + int(offset)

    return []

print(accum)
    
```
