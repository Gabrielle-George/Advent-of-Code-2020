```python
filename = "input.txt"
file = open(filename)

line = file.readline().strip()

problems = []
while line:
    problems.append(line)
    line = file.readline().strip()
```

```python
def multiply(a,b):
    return a*b
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def divide(a,b):
    return a/b
operators = {"*":multiply, "+":add,"-":subtract,"/":divide}
```

```python
prac = problems[0]
prac
```

```python
def solve_left_to_right(problem):
    cur = 0
    right = '-'
    left = '-'
    operator = ""
    while cur < len(problem):
        if problem[cur] in "+-/*":
            operator = problem[cur]
        elif problem[cur] == "(":
            paren_cur = cur+1
            #copy the innermost part
            paren_problem = []
            matching_paren = 1
            while matching_paren >0:
                if problem[paren_cur] =="(":
                    matching_paren = matching_paren+1
                if problem[paren_cur] ==')':
                    matching_paren = matching_paren-1
                if matching_paren >0:
                    paren_problem.append(problem[paren_cur])
                paren_cur = paren_cur +1
            paren_soln = solve_left_to_right(paren_problem)
            cur = paren_cur-1
            #check whether parenl_sln should be left or right
            if left == "-":
                left = paren_soln
            else:
                right = paren_soln
        elif left == '-':
            left = problem[cur]
        elif right =='-':
            right = problem[cur] 
        if right != '-' and operator !='-':
            print("{} {} {} = {}".format(left,operator, right,operators[operator](int(left),int(right))))
            left = operators[operator](int(left),int(right))
            right = '-'
            operator = '-'
        cur = cur+1
    return int(left)
        
        
        
        
```

```python
solve_left_to_right('((2 + 4 * 9) * (6 + 9 * 8 + 6) + 6) + 2 + 4 * 2'.replace(" ",""))

```

```python
solve_left_to_right('(2 + 3 + (8 + 9 + 4 * 8) * (5 * 3 * 7 + 9 * 5 * 8) + 6) * 9 + 6'.replace(" ",''))
```

```python
sums = 0
for i in problems:
    sums = sums + solve_left_to_right(i.replace(" ",""))
```
